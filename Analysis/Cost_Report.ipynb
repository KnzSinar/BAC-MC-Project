```python
#checking to see if the industry is profitable through net income

ni1= df1['Net Income'].sum()
ni2= df2['Net Income'].sum()
ni3= df3['Net Income'].sum()
ni4= df4['Net Income'].sum()
ni5= df5['Net Income'].sum()
ni6= df6['Net Income'].sum()
ni7= df7['Net Income'].sum()


nisum = pd.DataFrame({
    'Year': ['2015', '2016', '2017','2018','2019','2020','2021'],
     'Sum': [ni1,ni2,ni3,ni4,ni5,ni6,ni7],
})

#line plot
plt.figure(figsize=(10, 6))
plt.plot(nisum['Year'], nisum['Sum'], marker='o', color='blue', label='Net Income')
for x, y in zip(nisum['Year'], nisum['Sum']):
    plt.text(x, y, str(y), ha='center', va='bottom')

plt.title('Net Income Over the Years')
plt.xlabel('Year')
plt.ylabel('Net Income Sum')
plt.legend()
plt.grid(True)
plt.show()



#bar plot
plt.figure(figsize=(10, 6))
bars = plt.bar(nisum['Year'], nisum['Sum'], color='green')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), 
             str(int(bar.get_height())), ha='center', va='bottom')
    
plt.title('Net Income Over the Years')
plt.xlabel('Year')
plt.ylabel('Net Income Sum')
plt.show()

#from this we can understand that the industry is profitable however it is heavily affected by COVID
```

```python
#measuring the contribution towards the net income sum based on state
ni1 = df1['Net Income'].sum()
ni2 = df2['Net Income'].sum()
ni3 = df3['Net Income'].sum()
ni4 = df4['Net Income'].sum()
ni5 = df5['Net Income'].sum()
ni6 = df6['Net Income'].sum()
ni7 = df7['Net Income'].sum()

nisum = pd.DataFrame({
    'Year': ['2015', '2016', '2017', '2018', '2019', '2020', '2021'],
    'Sum': [ni1, ni2, ni3, ni4, ni5, ni6, ni7],
})

plt.figure(figsize=(30, 30))

for i in range(1, 8):
    year = 'df{}'.format(i)
    tyinc = vars()[year]['Net Income'].sum()
    state_income = {}
    for state_code, state_df in vars()[year].groupby('State Code'):
        state_income[state_code] = state_df['Net Income'].sum() / tyinc * 100
    
  
    ssinc = sorted(state_income.items(), key=lambda x: x[1], reverse=True)
    labels = [x[0] for x in ssinc]
    sizes = [x[1] for x in ssinc]
    
    plt.subplot(3, 3, i)
    plt.barh(labels, sizes, color='skyblue')
    plt.title('Contribution of States to Net Income - {}'.format(year))
    plt.xlabel('Percentage Contribution')
    plt.gca().invert_yaxis() 

plt.tight_layout()
plt.show()


#as we can see california has a very high contribution either if its positive or negative
```
