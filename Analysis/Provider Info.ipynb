```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/KnzSinar/BAC-MC-Project/main/Case%20Study%20Material/ProviderInfo_2015.csv', encoding='latin-1')
df2 = pd.read_csv('https://raw.githubusercontent.com/KnzSinar/BAC-MC-Project/main/Case%20Study%20Material/ProviderInfo_2016.csv', encoding='latin-1')
df3 = pd.read_csv('https://raw.githubusercontent.com/KnzSinar/BAC-MC-Project/main/Case%20Study%20Material/ProviderInfo_2017.csv', encoding='latin-1')
df4 = pd.read_csv('https://raw.githubusercontent.com/KnzSinar/BAC-MC-Project/main/Case%20Study%20Material/ProviderInfo_2018.csv', encoding='latin-1')
df5 = pd.read_csv('https://raw.githubusercontent.com/KnzSinar/BAC-MC-Project/main/Case%20Study%20Material/ProviderInfo_2019.csv', encoding='latin-1')
df6 = pd.read_csv('https://raw.githubusercontent.com/KnzSinar/BAC-MC-Project/main/Case%20Study%20Material/ProviderInfo_2020.csv', encoding='latin-1')
df7 = pd.read_csv('https://raw.githubusercontent.com/KnzSinar/BAC-MC-Project/main/Case%20Study%20Material/ProviderInfo_2021.csv', encoding='latin-1')
```

```python
c2k = ['provnum', 'PROVNAME', 'ADDRESS', 'CITY', 'STATE', 'ZIP', 'PHONE', 'COUNTY_SSA', 'COUNTY_NAME', 'OWNERSHIP', 'BEDCERT', 'RESTOT', 'CERTIFICATION', 'INHOSP', 'LBN', 'PARTICIPATION_DATE', 'overall_rating']


df1=df1.loc[:, c2k]
df2=df2.loc[:, c2k]
df3=df3.loc[:, c2k]

df1.info()
```

```python
c2k1 = ['PROVNUM', 'PROVNAME', 'ADDRESS', 'CITY', 'STATE', 'ZIP', 'PHONE', 'COUNTY_SSA', 'COUNTY_NAME', 'OWNERSHIP', 'BEDCERT', 'RESTOT', 'CERTIFICATION', 'INHOSP', 'LBN', 'PARTICIPATION_DATE', 'OVERALL_RATING']

df4=df4.loc[:, c2k1]
df4.info()
```

```python
c2k2 =['PROVNUM', 'PROVNAME', 'ADDRESS', 'CITY', 'STATE', 'ZIP', 'PHONE', 'COUNTY_SSA', 'County_name', 'OWNERSHIP', 'BEDCERT', 'RESTOT', 'CERTIFICATION', 'INHOSP', 'LBN', 'PARTICIPATION_DATE', 'Overall_Rating', 'ABUSE_ICON'
]

df5=df5.loc[:, c2k2]

df5.info()
```

```python
c2k3=[
    'Federal Provider Number', 'Provider Name', 'Provider Address', 'Provider City',
    'Provider State', 'Provider Zip Code', 'Provider Phone Number', 'Provider SSA County Code',
    'Provider County Name', 'Ownership Type', 'Number of Certified Beds',
    'Average Number of Residents per Day', 'Provider Type', 'Provider Resides in Hospital',
    'Legal Business Name', 'Date First Approved to Provide Medicare and Medicaid Services',
    'Overall Rating', 'Abuse Icon'
]


df6=df6.loc[:, c2k3]
df7=df7.loc[:, c2k3]

df6.info()
```

```python
#since abuse icons was introduced at 2019 any datasets prior would not have it


df1['Abuse Icon'] = 'N/A'
df2['Abuse Icon'] = 'N/A'
df3['Abuse Icon'] = 'N/A'
df4['Abuse Icon'] = 'N/A'




df1.rename(columns=dict(zip(df1.columns, c2k3)), inplace=True)
df2.rename(columns=dict(zip(df2.columns, c2k3)), inplace=True)
df3.rename(columns=dict(zip(df3.columns, c2k3)), inplace=True)
df4.rename(columns=dict(zip(df4.columns, c2k3)), inplace=True)
df5.rename(columns=dict(zip(df5.columns, c2k3)), inplace=True)
```

```python
#after deciding hawaii checking on abuse icon
import seaborn as sns

hidf = df7[df7['Provider State'] == 'HI']

unique_facility_names = hidf['Provider Name'].unique()
mask = hidf['Abuse Icon'] == 'Y'

plt.figure(figsize=(10, 15))
ax = sns.heatmap(mask.to_frame(), cmap='coolwarm', cbar=False, annot=False)
ax.set_yticks(range(len(unique_facility_names)))
ax.set_yticklabels(unique_facility_names)

plt.title('Presence of Abuse Icon in Hawaii')
plt.xlabel('')
plt.ylabel('Facility Name')
plt.show()

```
