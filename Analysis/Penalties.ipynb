```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/KnzSinar/BAC-MC-Project/main/Case%20Study%20Material/Penalties_2015.csv', encoding='latin-1')
df2 = pd.read_csv('https://raw.githubusercontent.com/KnzSinar/BAC-MC-Project/main/Case%20Study%20Material/Penalties_2016.csv', encoding='latin-1')
df3 = pd.read_csv('https://raw.githubusercontent.com/KnzSinar/BAC-MC-Project/main/Case%20Study%20Material/Penalties_2017.csv', encoding='latin-1')
df4 = pd.read_csv('https://raw.githubusercontent.com/KnzSinar/BAC-MC-Project/main/Case%20Study%20Material/Penalties_2018.csv', encoding='latin-1')
df5 = pd.read_csv('https://raw.githubusercontent.com/KnzSinar/BAC-MC-Project/main/Case%20Study%20Material/Penalties_2019.csv', encoding='latin-1')
df6 = pd.read_csv('https://raw.githubusercontent.com/KnzSinar/BAC-MC-Project/main/Case%20Study%20Material/Penalties_2020.csv', encoding='latin-1')
df7 = pd.read_csv('https://raw.githubusercontent.com/KnzSinar/BAC-MC-Project/main/Case%20Study%20Material/Penalties_2021.csv', encoding='latin-1')


df5.drop(columns='Processing Date',inplace=True)
df6.drop(columns='Processing Date',inplace=True)
df7.drop(columns='Processing Date',inplace=True)
```

```python
#keeping wanted columns
c2k = ['provnum', 'provname', 'address', 'city', 'state', 'zip', 'pnlty_date', 'pnlty_type', 'fine_amt', 'payden_strt_dt', 'payden_days']
df1=df1.loc[:, c2k]
df2=df2.loc[:, c2k]
df3=df3.loc[:, c2k]
df4=df4.loc[:, c2k]

```

```python
#keeping wanted columns
c2k2 = ['Federal Provider Number', 'Provider Name', 'Provider Address', 'Provider City', 'Provider State', 'Provider Zip Code', 'Penalty Date', 'Penalty Type', 'Fine Amount', 'Payment Denial Start Date', 'Payment Denial Length in Days']



df5=df5.loc[:, c2k2]
df6=df5.loc[:, c2k2]
df7=df5.loc[:, c2k2]
```

```python
#renaming columns
df1.rename(columns=dict(zip(df1.columns, c2k2)), inplace=True)
df2.rename(columns=dict(zip(df2.columns, c2k2)), inplace=True)
df3.rename(columns=dict(zip(df3.columns, c2k2)), inplace=True)
df4.rename(columns=dict(zip(df4.columns, c2k2)), inplace=True)
df5.rename(columns=dict(zip(df5.columns, c2k2)), inplace=True)
df6.rename(columns=dict(zip(df6.columns, c2k2)), inplace=True)
df7.rename(columns=dict(zip(df7.columns, c2k2)), inplace=True)
```

```python
#finding out how many payment denials there are by state 2018


countpd = df4.groupby('Provider State')['Penalty Type'].count()

sortctpd = countpd.sort_values(ascending=False)

st = sortctpd.index
cts = sortctpd.values

plt.figure(figsize=(20, 6))
plt.bar(st,cts)
plt.title('Count of Citations by State')
plt.xlabel('State')
plt.ylabel('Citations')
plt.show()
```

```python
#after deciding Hawaii

hidf = df7[df7['Provider State'] == 'HI']


countpd = hidf[hidf['Penalty Type'] == 'Payment Denial'].groupby('Provider Name').size()
countpd = countpd.reindex(hidf['Provider Name'].unique(), fill_value=0)
countpd = countpd.sort_values(ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(countpd.index, countpd.values, color='blue')
plt.title('Count of Payment Denials in Hawaii (Ascending Order)')
plt.xlabel('Provider Name')
plt.ylabel('Count')
plt.show()
```
